```python
import pandas as pd
import matplotlib as plt

tokyo = pd.read_csv("train/Tokyo.csv")
kyushu = pd.read_csv("train/Kyushu.csv")
chukyo = pd.read_csv("train/Chukyo.csv")
higa = pd.read_csv("train/Higashisurugawan.csv")
kinki = pd.read_csv("test.csv")
dataspe = pd.read_csv("Data_Specification.csv")
```

```python
dataspe.iloc[[49],[3]] = 'To Work'
tokyo_oc = dataspe[(dataspe.City == 'Tokyo') & (dataspe.Type == 'Occupation')][['Description','Code']]
kyushu_oc = dataspe[(dataspe.City == 'Kyushu') & (dataspe.Type == 'Occupation')][['Description','Code']]
chukyo_oc = dataspe[(dataspe.City == 'Chukyo') & (dataspe.Type == 'Occupation')][['Description','Code']]
higa_oc = dataspe[(dataspe.City == 'Higashisuruga') & (dataspe.Type == 'Occupation')][['Description','Code']]
kinki_oc = dataspe[(dataspe.City == 'Kinki') & (dataspe.Type == 'Occupation')][['Description','Code']]
#[['Code','Description']]
```

```python
print(dataspe[(dataspe.City == 'Tokyo') & (dataspe.Type == 'Trip_type')])
```

```python
print(dataspe[(dataspe.City == 'Kinki') & (dataspe.Type == 'Trip Type')])
```

```python
def rep_oc(df,dataspe):
    codelist = list(dataspe['Code'])
    codelist = [int(i) for i in codelist]
    deslist = list(dataspe['Description'])
    df['Occupation'].replace(codelist,deslist,inplace=True)
    return df
tokyo = rep_oc(tokyo,tokyo_oc)
kyushu = rep_oc(kyushu,kyushu_oc)
chukyo = rep_oc(chukyo,chukyo_oc)
higa = rep_oc(higa,higa_oc)
kinki = rep_oc(kinki,kinki_oc)
print(kinki)
```

```python
tokyo_tt = dataspe[(dataspe.City == 'Tokyo') & (dataspe.Type == 'Trip Type')][['Description','Code']]
kyushu_tt = dataspe[(dataspe.City == 'Kyushu') & (dataspe.Type == 'Trip Type')][['Description','Code']]
chukyo_tt = dataspe[(dataspe.City == 'Chukyo') & (dataspe.Type == 'Trip Type')][['Description','Code']]
higa_tt = dataspe[(dataspe.City == 'Higashisuruga') & (dataspe.Type == 'Trip Type')][['Description','Code']]
kinki_tt = dataspe[(dataspe.City == 'Kinki') & (dataspe.Type == 'Trip Type')][['Description','Code']]
def rep_tt(df,dataspe):
    codelist = list(dataspe['Code'])
    codelist = [int(i) for i in codelist]
    deslist = list(dataspe['Description'])
    df['Trip_type'].replace(codelist,deslist,inplace=True)
    return df
tokyo = rep_tt(tokyo,tokyo_tt)
kyushu = rep_tt(kyushu,kyushu_tt)
chukyo = rep_tt(chukyo,chukyo_tt)
higa = rep_tt(higa,higa_tt)
kinki = rep_tt(kinki,kinki_tt)
```

```python
#kinki_oc.merge(toyko_oc)
kinki_oc = kinki_oc.rename(columns = {'Code':'Code_kinki'})
tokyo_oc = tokyo_oc.rename(columns = {'Code':'Code_tokyo'})
kyushu_oc = kyushu_oc.rename(columns = {'Code':'Code_kyushu'})
chukyo_oc = chukyo_oc.rename(columns = {'Code':'Code_chukyo'})
higa_oc = higa_oc.rename(columns = {'Code':'Code_higa'})


ocmerge1 = kinki_oc.merge(tokyo_oc,on='Description',how='outer')
ocmerge2 = ocmerge1.merge(kyushu_oc,on='Description',how='outer')
ocmerge3 = ocmerge2.merge(chukyo_oc,on='Description',how='outer')
ocmerge4 = ocmerge3.merge(higa_oc,on='Description',how='outer')
print(ocmerge4)

ocmerge4.to_csv("_assets/merge_specifiaction_v2.csv")
```

```python
kinki_tt = kinki_tt.rename(columns = {'Code':'Code_kinki'})
tokyo_tt = tokyo_tt.rename(columns = {'Code':'Code_tokyo'})
kyushu_tt = kyushu_tt.rename(columns = {'Code':'Code_kyushu'})
chukyo_tt = chukyo_tt.rename(columns = {'Code':'Code_chukyo'})
higa_tt = higa_tt.rename(columns = {'Code':'Code_higa'})


ttmerge1 = kinki_tt.merge(tokyo_tt,on='Description',how='outer')
ttmerge2 = ttmerge1.merge(kyushu_tt,on='Description',how='outer')
ttmerge3 = ttmerge2.merge(chukyo_tt,on='Description',how='outer')
ttmerge4 = ttmerge3.merge(higa_tt,on='Description',how='outer')
print(ttmerge4)

ttmerge4.to_csv("_assets/merge_tt.csv")
```
